# Preprocessing

```python
# ====================== Loading Data from Alpha Vantage ==================================

api_key = '4UEMMQVKNE9JLN9S'

# American Airlines stock market prices
ticker = "AAL"

# JSON file with all the stock market data for AAL from the last 20 years
url_string = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=%s&outputsize=full&apikey=%s"%(ticker,api_key)

# Save data to this file
file_to_save = 'stock_market_data-%s.csv'%ticker

# If you haven't already saved data,
# Go ahead and grab the data from the url
# And store date, low, high, volume, close, open values to a Pandas DataFrame
if not os.path.exists(file_to_save):
    with urllib.request.urlopen(url_string) as url:
        data = json.loads(url.read().decode())
        # extract stock market data
        data = data['Time Series (Daily)']
        df = pd.DataFrame(columns=['Date','Low','High','Close','Open','Volume'])
        for k,v in data.items():
            date = dt.datetime.strptime(k, '%Y-%m-%d')
            data_row = [date.date(),float(v['3. low']),float(v['2. high']),
                        float(v['4. close']),float(v['1. open']),float(v['5. volume'])]
            df.loc[-1,:] = data_row
            df.index = df.index + 1
    print('Data saved to : %s'%file_to_save)        
    df.to_csv(file_to_save)

# If the data is already there, just load it from the CSV
else:
    print('File already exists. Loading data from CSV')
    df = pd.read_csv(file_to_save)
```

```python
df.drop(columns='Unnamed: 0',inplace=True)
```

```python
df['Mid'] = (df.High+df.Low)/2.
```

# Charting

```python
block_size = 60*3
models = []
count = 0
for block in range(block_size,len(df),15):
    
    training = df[block:block+block_size]
    stepwise_fit = auto_arima(training.Mid,
                              error_action='ignore',  # don't want to know if an order does not work
                              suppress_warnings=True,  # don't want convergence warnings
                              stepwise=True)  # set to stepwise

    periods_to_predict = 30
    predictions = stepwise_fit.predict(n_periods=periods_to_predict)
    models.append(stepwise_fit.get_params())
    
    predictions = pd.DataFrame({'predictions':predictions})
    new_index = range(training.index[-1],training.index[-1]+len(predictions))
    predictions.index=new_index
    plt.plot(training.Mid)
    plt.plot(predictions.predictions, linewidth=3)
    
    count+=1
    
    if count == 100:
        break
    #break
```

# Profit Testing

```python
def calculate_profit(buys,sells):
    
    if len(buys)>len(sells):
        profit = sum(sells)-sum(buys[:-1])
    else:
        profit = sum(sells)-sum(buys)
        
    return profit
```

```python
buys = []
sells = []
in_market = False
block_size = 60*6
buy_price = 0
count = 0
periods_to_predict = 20

for block in range(block_size,len(df)-block_size,10):
    
    training = df[block:block+block_size]
    stepwise_fit = auto_arima(training.Mid,
                              error_action='ignore',  # don't want to know if an order does not work
                              suppress_warnings=True,  # don't want convergence warnings
                              stepwise=True)  # set to stepwise

    predictions = stepwise_fit.predict(n_periods=periods_to_predict)
    prediction_gain = predictions[-1] - predictions[0]
    fee = predictions[-1]*.003 + predictions[0]*.003
    
    if in_market == False:
        if prediction_gain > fee:
            buy_price = training.Mid.iloc[-1]
            buys.append(buy_price)
            in_market = True
            in_market_minutes = 0
    if in_market == True:
        current_price = training.Mid.iloc[-1]
        profit = current_price - buy_price
        in_market_minutes+=1
        
        if profit > buy_price*.005:
            sells.append(current_price)
            in_market = False
        if in_market_minutes > periods_to_predict:
            if (prediction_gain < -fee):
                sells.append(current_price)
                in_market = False
                
    '''count+=1
    if count == 100:
        break'''
```

```python
calculate_profit(buys,sells)
```
